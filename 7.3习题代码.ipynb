```python
# WRF数据目录
wrfout_path = '/code/dataset/wrfout/'
```

```python
# 导入模块
import metpy.calc as mpcalc

from netCDF4 import Dataset
from wrf import getvar, uvmet

wrf_file = Dataset(wrfout_path + 'wrfout_d01_2019-08-08_18_00_00')

# 提取 WRF 模拟的三维风场，通过[0]进行索引仅提取其中一个高度层的uv风速
ua = getvar(wrf_file, 'ua', timeidx=0)[0]
va = getvar(wrf_file, 'va', timeidx=0)[0]

# 根据 uv 分量计算风速
ws = mpcalc.wind_speed(ua, va)
ws.max()
```
